## Analysis 2: Adj_Volume and its moving average in one plot
Generate volume bar chart and add its moving average in lines. if close price is greater than open price, the volume bar that day is green, otherwise, it's red.

```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import os
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib.dates as mdate
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

from datetime import datetime
```

```python
# Read downloaded data
p = Path(os.getcwd())
path = str(p.parent) + '/data/AAPL/AAPL.csv'
df = pd.read_csv(path)
```

```python
df = df[df.apply(lambda x: int(x['Date'].split('-')[0]) >= 2016, axis=1)]
df['Date'] = df['Date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d').date())
#df = df.set_index('Date')
df = df[0:30]
df = df[df.apply(lambda x: int(x['Adj_Volume'])> 0, axis=1)]
df = df.reindex(index=df.index[::-1])
df.head()
```

```python
fig, ax = plt.subplots()
ax = df['Adj_Volume'].plot(kind='Bar', color=['g' if x[5] > x[2] else 'r' for x in df.itertuples()])
ax.set_xticklabels([x.strftime('%Y-%m-%d') for x in df['Date']])
fig.set_size_inches(10,4)
```

```python
# moving average
moving_average_day = [2,5,10]

for ma in moving_average_day:
    column_name = "MA for %s days" %(str(ma))
    df[column_name]=Series.rolling(df['Adj_Volume'],ma).mean()
#df.head()
```

```python
fig, ax = plt.subplots()
for ma in (moving_average_day):
    df[["MA for %s days" %(str(ma))]].plot(kind='line', ax=ax, use_index=False)
df[['Volume']].plot(kind='bar', ax=ax, color=['g' if x[5] > x[2] else 'r' for x in df.itertuples()])  
ax.set_xticklabels([x.strftime('%Y-%m-%d') for x in df['Date']])
fig.set_size_inches(10,4)
plt.savefig('ana_2_output/volume+moving_average.png')
```
