## Use NYT API to collect NYT data. Perform 3 analysis on the collected data.
Anal1: use archive data from 2015-2016, count top 50 word frequently used in abstracts

```python
import glob
import string
import os
import nltk
import json
from pathlib import Path
from collections import Counter
```

```python
p = Path(os.getcwd())
path = str(p.parent) + '/data/Archive/*/*.txt'
files = glob.glob(path)
#print(path)
```

```python
abstracts = []
for file in files:
    with open(file, 'r', encoding='utf-8', errors='ignore') as f:
        python_data = json.load(f)
        docs = python_data['response']['docs']
        for doc in docs:
            if 'abstract' in doc and doc['abstract'] is not None:
                abstract = doc['abstract']
                abstracts.append(doc['abstract'])
print(abstracts[0:10])
```

```python
stop_words = set(nltk.corpus.stopwords.words('english'))
word_counter = Counter()
for abstract in abstracts:
    for word in map(lambda w: w.strip(string.punctuation).lower(), abstract.split(' ')):
        if word != '' and word not in stop_words and word.isalpha():
            word_counter.update([word])
#print(len(word_counter))
```

```python
#print(word_counter)
```

```python
import csv
output_path = 'anal1Output/word_Count_abstract.csv'
with open(output_path, 'w') as outcsv:
        writer = csv.DictWriter(outcsv, fieldnames = ["WORD","FREQUENCY"])
        writer.writeheader()
        for word,freq in word_counter.most_common(50):
            writer.writerow({'WORD': word,'FREQUENCY': str(freq)})
```
